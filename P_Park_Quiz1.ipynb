**Definition:** A diamond is a precious stone consisting of a clear and colorless crystalline form of pure carbon. It is the hardest naturally occurring substance on the planet. 

**Analysis:** With a rich history of prestige and value, diamonds have been sought after throughout history. This revered stone is still considered the crowning gem and consequently the attributes of a diamond are highly scrutinized. There are 10 distinct attributes that are analyzed with regards to the value of a diamond. 

These are 10 distinct attributes:

1. carats
2. cut
3. color
4. clarity
5. depth
6. table
7. price
8. length in mm (0--10.74)
9. width in mm (0--58.9)
10. depth in mm (0--31.8)

According to The Diamond Pro, the "4 C's" are the crème de la crème of all the major attributes (Please refer to: https://www.diamonds.pro/education/4cs-diamonds/).

The **cut** is considered the premier attribute and experts look for refinement and immaculate precision. Please note that the scale in the dataset is as follows (the parentheticals corresponds to the matching term in the visual below): Premium (excellent), Ideal (very good), Very Good (good), Good (fair), Fair (poor).

The **color** is considered the 2nd most important attribute, and high-level diamonds should be white or colorless. 

The **clarity** is third; exceptional diamonds should be blemish-free and there should be nothing taking away from the brilliance of the diamond. 

Fourth, is the **carat**; although size matters, it is a matter of preference to the wearer and this characteristic does not supersede the previous three. 

Here is a visual breakdown of the high-to-low standards of the "4 C's." (Please refer to https://4cs.gia.edu/en-us/blog/gia-diamond-grading-scales/):

Here are the other evaluative attributes:

**Depth**

The depth of a diamond is the length measured from the bottom tip to the flat, top surface; otherwise known as the table. 

Please refer to: https://www.withclarity.com/education/diamond-education/diamond-cut/what-is-diamond-depth-or-diamond-education#:~:text=Diamond%20depth%20is%20a%20crucial,depth%20of%20diamond%20in%20millimeters.


**Table**

The table of the diamond is the stone that is visible as it is situated on the wearer's finger. It is the largest component on a diamond and is essentially the visible aesthetic of the diamond from the perspective of the viewers. It is also the component that reflects light and thus it gives off the diamond's brilliance. 

Please refer to: https://www.stevenstone.co.uk/expert-advice/table-depth#:~:text=A%20diamond's%20table%20is%20the,reflected%20back%20from%20the%20pavilion.


**Price**

A diamond's value is essentially based on the "4 C's." This is comprised of cut, color, clarity, and carat. Appraisers use these four attributes when placing a value on a diamond. The secondary attributes include: table, depth, and the length, width, and depth of the diamond. 


Please refer to: https://www.withclarity.com/blog/2020/07/22/estimating-diamond-value/#:~:text=A%20diamond's%20value%20is%20determined,the%20quality%20of%20the%20band.

**Length, Width and Depth (in mm)** 

The length of the diamond ranges from 0 - 10.74 mm.

The width of the diamond ranges from 0 - 58.9 mm.

The depth of the diamond ranges from 0 - 31.8 mm.

***Importing Libraries and Reading the Data Set***

We need to import the necessary libraries and assign it with an alias.
The variable **data** is assigned to pd.read_csv("https://raw.githubusercontent.com/philipjpark/P_Park_Quiz1/main/diamonds.csv"). Now we can call on data and read the information. <B> <big> Exploratory Data Analysis </big> </B> 

**1. Carats:**

I decided to use a histogram to show ranges of carats to compare the highest levels of frequency for carat size. The highest level of frequency ranges from (0.2-0.3) to (0.5-0.6). 

The frequencies ranging from 0-1 and 1-2 make up the vast majority of the diamonds in this dataset. The inference that can be made is that diamonds with a carat size of 2+ is difficult to acquire because of its correlation to an increasing price. 

```python
Tdata= data['carat']
print(Tdata.head())
plt.figure(figsize=(10,10))
plt.hist(Tdata,bins=np.arange(0, 5, .10),alpha=0.99)
plt.title("Carats", fontsize=20, color='white')
plt.xlabel('Carat Size', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.show()
```

Here is a closer look at the frequencies of diamonds that range from a carat size of 2 - 5:



```python
Tdata= data['carat']
print(Tdata.head())
plt.figure(figsize=(8,8))
plt.hist(Tdata,bins=np.arange(2, 5, .10),alpha=0.96)
plt.title("Carats", fontsize=20, color='white')
plt.xlabel('Carat Size', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.show()
```

**2. Cut:**

The cut refers to the quality of the cut. The rankings from the lowest to the highest quality is as follows: Fair, Good, Very Good, Ideal, and Premium. The pie graph indicates that cut qualities of diamonds are frequently ideal or premium. 

This supports the claim that the cut is the most important attribute of a diamond. Generally, concessions will not be made with regards to cut quality.  

```python
plt.figure(figsize=(11,11))
plt.title("Cut", fontsize=20, color='white')
plt.ylabel('Frequency')
data.groupby('cut').size().plot(kind='pie')
```

**3. Color:**

The color is ranked from D to J. D is the highest ranking and J is the lowest ranking. The letters decrease in ranking as they gets closer to J. Starting with D, if we go around the pie graph counterclockwise, we would be going in the direction of decreasing ranking. D - H have frequencies that are similar. I and J, which are the lowest ranking, have the lowest frequencies. 

The highest ranking of D means that it is essentially colorless. Diamond rankings can go as low as Z, which has a light yellow to brown hue. Please refer to: https://www.diamonds.pro/education/color/#:~:text=Diamond%20color%20is%20measured%20using,near%20colorless%20to%20slightly%20yellow.

```python
plt.figure(figsize=(11,11))
plt.title("Color", fontsize=20, color='white')
plt.ylabel('Frequency')
data.groupby('color').size().plot(kind='pie')
```

***4. Clarity:***

Clarity is a measure of how clear the diamond is. The ranking is as follows: I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best). The count plot clearly shows which clarity rankings are most frequent in this dataset. 

Similar to the other "4 C's", clarity is an attribute that few concessions will be made. However, the distribution is not as rigid as it is for carats and cut. This may give credence to the notion that clarity is a clear third when it comes to the hierarchy of diamond attributes. 

```python
plt.figure(figsize=(11,11))
plt.title("Clarity", fontsize=20, color='white')
sns.countplot(data=data, x='clarity', order=data.clarity.value_counts().index)
```

***5. Depth:***

The total depth percentage ranges from 43 to 79% and the depth percentage that is most prevalent ranges from 61 - 63%. Further research into this distribution should shed more light on the significance of the 61 - 63% total depth percentage. 

```python
Tdata= data['depth']
print(Tdata.head())
plt.figure(figsize=(10,10))
plt.hist(Tdata,bins=np.arange(43, 79, .5), alpha=0.8)
plt.title("Total Depth Percentage", fontsize=20, color='white')
plt.xlabel('Percentage')
plt.ylabel('Frequency')
plt.show()
```

***6. Table:***

Table refers to the width of the top of the diamond relative to its widest point. It ranges from 43 - 95 mm. The highest range is approximately 56-57 and it approaches a density of 0.30. 

```python

plt.figure(figsize=(9,9))
sns.distplot(data.table, bins =20, color="yellow")
plt.style.use("dark_background")
plt.title("Ranges of Table Size", fontsize=20, color='white')
plt.show()
```

**Problem #1**

What proportion of diamonds are between .30 and 1.08 carats?

**Answer:** Approximately 75% of the total diamonds have a carat size between .30 and 1.08 carats. 

```python
prob = (data[(data['carat']>= 0.3) & (data['carat']<= 1.08)].shape[0])/(data.shape[0])
percent ="{:.0%}".format(prob)
percent
```

***Problem #2***

How many of the diamonds have equal x and y dimensions?


**Answer:** 17 diamonds have the same dimensions for its length and width. 

```python
lenEqualsWidth = data[data['x'] == data['y']].shape[0]

lenEqualsWidth
```

***Problem #3***

How many of the diamonds have a carat value that is less than the mean carat value?

**Answer:** 31,127 diamonds have a carat value less than the mean carat value. 

```python
data[data['carat'] < data['carat'].mean()].shape[0]
```

***Problem #4***

How many diamonds have a Premium cut or better? Note that the possible values for the quality of a cut are ranked in ascending order as follows: Fair/ Good / Very Good / Premium / Ideal

**Answer:** There are 35,342 diamonds that have a Premium cut or better. 

```python
data[(data['cut'] == "Premium") | (data['cut'] == 'Ideal')].shape[0]
```

***Problem #5*** 

Which diamond has the highest price per carat? What is its value?

**Answer:** Diamond# 27636 has the highest price per carat. The value is $17,828.85

```python
store = data
store['per carat'] = (store['price']/store['carat'])

highest = store[store['per carat'] == store['per carat'].max()]

print('Dollars per carat: %.2f .', round(highest['per carat'], 2))
print('Price is: $',highest['price'])
highest
```
